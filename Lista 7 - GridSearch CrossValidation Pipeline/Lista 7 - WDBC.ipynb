```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
# models
from sklearn.linear_model import Perceptron
```

```python
df = pd.read_csv('../Datasets/wdbc.data', header=None)
df.head()
```

```python
df.info()
```

##### Excluindo coluna ID e Passando o Label para a ultima posição:

```python
df.drop(columns=0, inplace=True)
df.head()
```

```python
label = df[1]
df.drop(columns=1, inplace=True)
df['label'] = label
df.head()
```

```python
df.isna().sum() # Não tem valores faltantes
```

```python
df = pd.get_dummies(df ,prefix='label', drop_first=True)
df.head()
```

```python
X = df.values[:, :-1]
y = df.values[:,  -1]
```

```python
#na verdade é para ser usado o stratfied_shuffle_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, stratify=y, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Usando o GridSearch

```python
parametros = {'estimator__penalty':['None', 'l1', 'l2', 'elasticnet'], 'estimator__alpha':[.000001, .00001, .0001, .001, .01, .1, 1]}

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

pipeline = Pipeline([('transformer', StandardScaler()), ('estimator', Perceptron(max_iter=1000, tol=1e-3))])

gs = GridSearchCV(pipeline, parametros, cv=cv, scoring='accuracy', iid=True, return_train_score=True)
gs.fit(X_train, y_train);
```

```python
cols = ['param_estimator__alpha','param_estimator__penalty','mean_train_score','mean_test_score']
pd.DataFrame(gs.cv_results_)[cols].sort_values(by='mean_test_score', ascending=False)
```

Os melhores modelos foram o Perceptron com alpha=0.00001 usando a penalidade l2 e o mesmo usando a penalidade elasticnet

```python
##### Standartizando os dados para o mododelo
final_scaler = StandardScaler()
X_train_std = final_scaler.fit_transform(X_train)
X_test_std = final_scaler.transform(X_test)
```

```python
model1 = Perceptron(max_iter = 1000 , tol = 1e-3, penalty='elasticnet', alpha=.00001)
model1.fit(X_train_std, y_train)
y_pred1 = model1.predict(X_test_std)
accuracy_score(y_test, y_pred1)
```

```python
model2 = Perceptron(max_iter = 1000, tol = 1e-4, penalty='l2', alpha=.0001)
model2.fit(X_train_std, y_train)
y_pred2 = model2.predict(X_test_std)
accuracy_score(y_test, y_pred2)
```

#### Notamos que mesmo usando o dataset de treino inteiro (diferentemente da crossvalidation) ele não conseguiu obter uma acuraccy melhor, ps: o conjunto de teste era maior.
