# KNN

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../Datasets/iris-dataset.csv', header=None) # lendo o dataset
df.head()
```

### Mapenado o Label

```python
mapa = {'Iris-setosa':0,'Iris-versicolor':1, 'Iris-virginica':2}
df[4] = df[4].map(mapa)
```

```python
df.isna().sum() #Verificando se existem valores faltantes
```

```python
X = df.values[:, :-1] # Separando os dados
y = df.values[:, -1]
```

# Apresentando os dados

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

img = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=X[:, 3], cmap=plt.viridis())
fig.colorbar(img);
```

```python
class KNNmodel:
    
    dist = None
    
    def __init__(_self, dist='euclidean'):
        _self.dist = dist
        
    def KNNClassifier(_self, X, y, ponto, k, p=2):
        knn_id = _self.id_KNN(X, ponto, k, p)
        count = np.bincount(np.int64(y[knn_id]))
        return np.argmax(count)
        
    def KNNRegressor(_self, X, y, ponto, k, p=2):
        knn_id = _self.id_KNN(X, ponto, k)
        return np.mean(X[knn_id])
    
    def minkowski_distance(_self, X, ponto, p):
        return np.sum(np.abs(X - ponto) ** p, axis=1) ** (1/p)        
        
    def euclidean_distance(_self, X, ponto): # Dica: usar minkowski_distance com p=2.
        return np.sqrt( np.sum(((X - ponto)**2), axis=1) )

    def manhattan_distance(_self, X, ponto): # Dica: usar minkowski_distance com p=1.
        return np.sum(np.abs(X - ponto), axis=1)
        
    def chebyshev_distance(_self, X, ponto):
        return np.max(np.abs(X - ponto), axis=1) 
    
    def id_KNN(_self, X, ponto, k, p):
        medida = []
        if(_self.dist=='euclidean'):
            medida = _self.euclidean_distance(X, ponto)
        elif(_self.dist=='minkowski'):
            medida = _self.minkowski_distance(X, ponto, p)
        elif(_self.dist=='manhattan'):
            medida = _self.manhattan_distance(X, ponto)
        else:
            medida = _self.chebyshev_distance(X, ponto)
        ordem_id = np.argsort(medida)
        return ordem_id[:k]
```

# Dividindo o dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.9, stratify=y, random_state=42)
```

# Standartizando os dados 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

# Criando modelos

```python
from sklearn.metrics import accuracy_score
```

```python
k = [1, 3, 5]
```

##### Euclidean

```python
euclidean = KNNmodel(dist='euclidean')
```

```python
euclidean_pred = {}
for n in k:
    euclidean_pred[n] = []
    for i in range(X_test_std.shape[0]):
        euclidean_pred[n].append(euclidean.KNNClassifier(X_train_std, y_train, X_test_std[i], n))
    print('K = ' + str(n) + ' -> ' + str(accuracy_score(euclidean_pred[n], y_test)))
```

```python
minkowski = KNNmodel(dist='minkowski')
```

```python
minkowski_pred = {}
for n in k:
    minkowski_pred[n] = []
    for i in range(X_test_std.shape[0]):
        minkowski_pred[n].append(minkowski.KNNClassifier(X_train_std, y_train, X_test_std[i], n, p=4))
    print('K = ' + str(n) + ' -> ' + str(accuracy_score(minkowski_pred[n], y_test)))
```

```python
manhattan = KNNmodel(dist='manhattan')
```

```python
manhattan_pred = {}
for n in k:
    manhattan_pred[n] = []
    for i in range(X_test_std.shape[0]):
        manhattan_pred[n].append(manhattan.KNNClassifier(X_train_std, y_train, X_test_std[i], n, p=15))
    print('K = ' + str(n) + ' -> ' + str(accuracy_score(manhattan_pred[n], y_test)))
```

```python
chebyshev = KNNmodel(dist='chebyshev')
```

```python
chebyshev_pred = {}
for n in k:
    chebyshev_pred[n] = []
    for i in range(X_test_std.shape[0]):
        chebyshev_pred[n].append(chebyshev.KNNClassifier(X_train_std, y_train, X_test_std[i], n, p=4))
    print('K = ' + str(n) + ' -> ' + str(accuracy_score(chebyshev_pred[n], y_test)))
```

# Comparando os resultados

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = {}
models['skeuclidean_k1'] = KNeighborsClassifier(n_neighbors= 1, metric='euclidean')
models['skeuclidean_k3'] = KNeighborsClassifier(n_neighbors= 3, metric='euclidean')
models['skeuclidean_k5'] = KNeighborsClassifier(n_neighbors= 5, metric='euclidean')
models['skminkowski_k1'] = KNeighborsClassifier(n_neighbors= 1, metric='minkowski', p=4)
models['skminkowski_k3'] = KNeighborsClassifier(n_neighbors= 3, metric='minkowski', p=4)
models['skminkowski_k5'] = KNeighborsClassifier(n_neighbors= 5, metric='minkowski', p=4)
models['skmanhattan_k1'] = KNeighborsClassifier(n_neighbors= 1, metric='manhattan')
models['skmanhattan_k3'] = KNeighborsClassifier(n_neighbors= 3, metric='manhattan')
models['skmanhattan_k5'] = KNeighborsClassifier(n_neighbors= 5, metric='manhattan')
models['skchebyshev_k1'] = KNeighborsClassifier(n_neighbors= 1, metric='chebyshev')
models['skchebyshev_k3'] = KNeighborsClassifier(n_neighbors= 3, metric='chebyshev')
models['skchebyshev_k5'] = KNeighborsClassifier(n_neighbors= 5, metric='chebyshev')
```

```python
y_preds = {}
for i in models:
    models[i].fit(X_train_std, y_train)
    y_preds[i] = models[i].predict(X_test_std)
    print(i + ' --> ' + str(accuracy_score(y_preds[i], y_test)))
```
