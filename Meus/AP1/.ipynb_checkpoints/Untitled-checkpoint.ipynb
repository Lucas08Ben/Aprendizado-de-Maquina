# Questão 3

```python
X = df.values[:, :-1]
y = df.values[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, stratify=y, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
models = {}
models['Perceptron'] = Perceptron(max_iter = 1000 , tol = 1e-3)
models['Perceptron_std'] = Perceptron(max_iter = 1000 , tol = 1e-3)
models['SGDClassifier'] = SGDClassifier(max_iter = 1000 , tol = 1e-3)
models['SGDClassifier_std'] = SGDClassifier(max_iter = 1000 , tol = 1e-3)
models['LogisticRegression'] = LogisticRegression(solver='lbfgs', max_iter = 10000)
models['LogisticRegression_std'] = LogisticRegression(solver='lbfgs', max_iter = 10000)
```

```python
y_pred_test = {}
y_pred_train = {}
for k in models:
    if k[-3:] == 'std':
        models[k].fit(X_train_std, y_train)
        y_pred_test[k] = models[k].predict(X_test_std)
        y_pred_train[k]= models[k].predict(X_train_std)
    else:
        models[k].fit(X_train, y_train)
        y_pred_test[k] = models[k].predict(X_test)
        y_pred_train[k]= models[k].predict(X_train)
```

# Questão 4

```python
def print_perc_classes(y):
    values, counts = np.unique(y, return_counts=True)
    total_size = y.shape[0]
    for i in range(values.shape[0]):
        print('class: {}, count: {}, perc: {}'.format(int(values[i]), counts[i], (counts[i]/total_size).round(4)))
```

```python
print_perc_classes(y)
```

```python
print_perc_classes(y_train)
```

```python
print_perc_classes(y_test)
```

Está estratificado pois as porcentagens do y_test e y_train estão muito parecidas com a porcentagem da y

# Questão 5

```python
def acc(y, y_pred):
    return sum(y == y_pred)/y.shape[0]
```

```python
accs_test = {}
accs_train = {}
for k in models:
    accs_train[k] = acc(y_test, y_pred_train[k])
    accs_test[k]  = acc(y_test, y_pred_test[k])
    print(k + '\t---> test: ' + str(accs_test[k]) +'\t|train: ' + str(accs_train[k]))
```
