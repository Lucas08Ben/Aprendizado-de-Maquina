# WDBC

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import accuracy_score

# models
from sklearn.linear_model import Perceptron
```

```python
df = pd.read_csv('wdbc.data', header=-1)
df.head()
```

```python
df.info()
```

##### Excluindo coluna ID e Passando o Label para a ultima posição:

```python
df.drop(columns=0, inplace=True)
df.head()
```

```python
label = df[1]
df.drop(columns=1, inplace=True)
df['label'] = label
df
```

```python
df.isna().sum() # Não tem valores faltantes
```

```python
df = pd.get_dummies(df ,prefix='label', drop_first=True)
df.head()
```

```python
X = df.values[:, :-1]
y = df.values[:,  -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, stratify=y, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
models = {}
models['Perceptron_01'] = Perceptron(max_iter = 1000 , tol = 1e-3, alpha=.01)
models['Perceptron_01_l1'] = Perceptron(max_iter = 1000 , tol = 1e-3, penalty='l1', alpha=.01)
models['Perceptron_01_l2'] = Perceptron(max_iter = 1000 , tol = 1e-3, penalty='l2', alpha=.01)
models['Perceptron_01_elasticnet'] = Perceptron(max_iter = 1000 , tol = 1e-3, penalty='elasticnet', alpha=.01)
models['Perceptron_0001'] = Perceptron(max_iter = 1000 , tol = 1e-3, alpha=.0001)
models['Perceptron_0001_l1'] = Perceptron(max_iter = 1000 , tol = 1e-3, penalty='l1', alpha=.0001)
models['Perceptron_0001_l2'] = Perceptron(max_iter = 1000 , tol = 1e-3, penalty='l2', alpha=.0001)
models['Perceptron_0001_elasticnet'] = Perceptron(max_iter = 1000 , tol = 1e-3, penalty='elasticnet', alpha=.0001)
```

```python
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
```

```python
accs = {}
for k in models:
    accs[k] = cross_val_score(models[k], X_train_std, y_train, cv=cv, scoring='accuracy')
    print(k + ':\t\t' + str(accs[k]) + '\t->>\t' + str(sum(accs[k])/3))
```

Os melhores modelos foram o Perceptron com alpha=0.0001 usando a penalidade l2 e o mesmo usando a penalidade elasticnet

```python
model1 = Perceptron(max_iter = 1000 , tol = 1e-3, penalty='elasticnet', alpha=.0001)
model1.fit(X_train_std, y_train)
y_pred1 = model1.predict(X_test_std)
acc1 = accuracy_score(y_test, y_pred1)
acc1
```

```python
model2 = Perceptron(max_iter = 1000 , tol = 1e-3, penalty='l2', alpha=.0001)
model2.fit(X_train_std, y_train)
y_pred2 = model2.predict(X_test_std)
acc2 = accuracy_score(y_test, y_pred2)
acc2
```
